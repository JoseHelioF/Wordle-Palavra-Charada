```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from unidecode import unidecode
from time import sleep, time
import pyautogui

def escrevePalavra(palavra):
    for letra in palavra:
        pyautogui.press(letra)
def pegaLetraQueNaoContem():
    resultado = []
    letras_erradas = driver.find_elements(By.CLASS_NAME, 'border-slate-400')
    for letra in letras_erradas:
        resultado.append(letra.text)
    letras_unicas = []
    [letras_unicas.append(letra) for letra in resultado if letra not in letras_unicas]
    return letras_unicas

def pegaLetraNaPosicaoErrada():
    resultado = []
    letras_posicao_erradas = driver.find_elements(By.CLASS_NAME, 'border-yellow-500')
    for letra in letras_posicao_erradas:
        resultado.append(letra.text)
    letras_unicas = []
    [letras_unicas.append(letra) for letra in resultado if letra not in letras_unicas]
    return letras_unicas

def retirarLetras(anterior,tirar):
    return [letra for letra in anterior if letra not in tirar]

def geraMatrizLetraFixa(array_resultante):
    matriz = []
    for i in range(0,len(array_resultante),5):
        matriz.append(array_resultante[i:i+5])
    return [next((parte[indice] for parte in matriz if parte[indice]), []) for indice in range(5)]

def geraVetorLetraFixa():
    elementos = driver.find_elements(By.CLASS_NAME, 'cellAnimation')
    array_resultante = []
    for elemento in elementos:
        if "border-green-500" in elemento.get_attribute("class"):
            array_resultante.append([elemento.text])
        else:
            array_resultante.append([])
    return geraMatrizLetraFixa(array_resultante)

def geraMatrizLetraPosicaoErrada(array_resultante):
    matriz = []
    for i in range(0,len(array_resultante),5):
        matriz.append(array_resultante[i:i+5])
    colunas_letras = [[] for _ in range(len(matriz[0]))]
    for coluna in matriz:
        for i, elemento in enumerate(coluna):
            if elemento:
                colunas_letras[i].extend(elemento)
#     print(colunas_letras) 
    return colunas_letras

def geraVetorLetraPosicaoErrada():
    elementos = driver.find_elements(By.CLASS_NAME, 'cellAnimation')
    array_resultante = []
    for elemento in elementos:
        if "border-yellow-500" in elemento.get_attribute("class"):
            array_resultante.append([elemento.text])
        else:
            array_resultante.append([])
#     print(array_resultante)
    return geraMatrizLetraPosicaoErrada(array_resultante)

def geraArrayComLetrasParaCombinar(letra_para_gerar_combinacao,array_letra_fora_posicao):
    resultado = []
    for letra in array_letra_fora_posicao:
        resultado.append([x for x in letra_para_gerar_combinacao if x not in letra])
    return resultado

def geraArrayDeCobinacao(array_letra_fixa,letras_para_combinar_cada_coluna):
    resultado = []
    for fixa,combinacao in zip(array_letra_fixa,letras_para_combinar_cada_coluna):
        if fixa:
            resultado.append(fixa)
        else:            
            resultado.append(combinacao)
    return resultado
def gerarCombinacao(array_contendo_combinacao):
    resultado = []
    for a in array_contendo_combinacao[0]:
        for b in array_contendo_combinacao[1]:
            for c in array_contendo_combinacao[2]:
                for d in array_contendo_combinacao[3]:
                    for e in array_contendo_combinacao[4]:
                        resultado.append(a+b+c+d+e)
    return resultado

def compararLista(base,palavras_resultado_da_combinacao):
    return [palavra for palavra in palavras_resultado_da_combinacao if palavra.lower() in base]

def encontrar_palavras_com_letra(letras, palavras):
    if letras:
        for letra in letras:
            resultado = [palavra for palavra in palavras if letra in palavra]
            palavras = resultado  
        return resultado
    else:
        return palavras
def clicarEnter():
    novo_jogo = driver.find_elements(By.CLASS_NAME, "bg-emerald-600")
    if novo_jogo:
        novo_jogo[0].click()
    botao_enter = driver.find_elements(By.XPATH, '//*[@id="root"]/div/div[3]/div[3]/button[8]')
    while len(botao_enter) == 0:
        botao_enter = driver.find_elements(By.XPATH, '//*[@id="root"]/div/div[3]/div[3]/button[8]')            
    botao_enter[-1].click()
#Variaveis 
alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Configurações do Chrome
opcoes = Options()
opcoes.add_argument("--start-maximized")
driver = webdriver.Chrome(options=opcoes)

# Abre o site
driver.get('https://charada.vercel.app/')
sleep(3)

#Pega a base de palavras
with open(r'C:\Users\joseh\Documents\Projeto\Python\PalavraCharada\charada.txt', mode='r') as entrada:
    base = entrada.read()  # Isso lê o conteúdo do arquivo
base = base.split('\n')
tempoTotal = 0
cont=0
estimativa = []
#*******************************************************************************************************************************
for loop_externo in range(5):
    ti = time()
#     palavra_base = ['rosea','multi','dance','pifao','beiju']
#     palavra_base = ['rosea','multi','dance','pifao']
    palavra_base = ['rosea','cupim','tango','veloz']
    if len(palavra_base)>5:
        print('Encolher menos palavras_base')
        break
    for palavra in palavra_base:
        escrevePalavra(palavra)
        
        clicarEnter()
    #***************************************************************************************************************************
    for loop_interno in range(5):
        letras_que_nao_contem = pegaLetraQueNaoContem()
        if len(letras_que_nao_contem)==0:
            loop_externo-=1
            print('Reiniciar o loop')
            break
        letras_na_posicao_errada = pegaLetraNaPosicaoErrada()
        letras_que_nao_contem = [letra for letra in letras_que_nao_contem if letra not in letras_na_posicao_errada]
        letra_para_gerar_combinacao = retirarLetras(alfabeto,letras_que_nao_contem)
        array_letra_fixa = geraVetorLetraFixa()
        array_letra_fora_posicao = geraVetorLetraPosicaoErrada()
        letras_para_combinar_cada_coluna = geraArrayComLetrasParaCombinar(letra_para_gerar_combinacao,array_letra_fora_posicao)
        array_contendo_combinacao = geraArrayDeCobinacao(array_letra_fixa,letras_para_combinar_cada_coluna)
        palavras_resultado_da_combinacao = gerarCombinacao(array_contendo_combinacao)
        palavras_com_letra_fora = encontrar_palavras_com_letra(letras_na_posicao_errada,palavras_resultado_da_combinacao)
        lista_comparada = compararLista(base,palavras_com_letra_fora)

        if len(lista_comparada) == 0:
            lista_comparada = ['teste']
        escrevePalavra(lista_comparada[0])
        clicarEnter()

        fim = driver.find_elements(By.CLASS_NAME, 'right-4')
        if fim:
            if driver.find_elements(By.CLASS_NAME, 'bg-green-100'):
                print('Acerto',lista_comparada[0])
            else:
                sleep(1)
                palavra_acrescentar = driver.find_elements(By.TAG_NAME, 'b')[0].text.lower()
                while palavra_acrescentar=='':
                    palavra_acrescentar = driver.find_elements(By.TAG_NAME, 'b')[0].text.lower()
                
                print('Erro, palavra certa é',palavra_acrescentar)
                if unidecode(palavra_acrescentar) not in base:
                    base.append(unidecode(palavra_acrescentar))
                    with open('charada.txt','w') as saida:
                        for palavra in base:
                            saida.write(palavra)
                            saida.write('\n')
                    print('Palavra ',palavra_acrescentar,' acrescentada')

            novo_jogo = driver.find_elements(By.CLASS_NAME, 'bg-emerald-600')
            novo_jogo[0].click()
            break

        palavra_errada = driver.find_elements(By.CLASS_NAME, 'bg-rose-200')
        if len(palavra_errada)>0:
            print('Palavra certa no jogo era',lista_comparada[0])
            botao_apaga = driver.find_elements(By.CLASS_NAME, 'text-base')
            if botao_apaga:
                for i in range(5):
                    botao_apaga[19].click()
                base = [palavra for palavra in base if palavra.upper() not in lista_comparada[0]]
            with open('charada.txt','w') as saida:
                for palavra in base:
                    saida.write(palavra)
                    saida.write('\n')
                
    tempoTotal += time() - ti
    print('loop_externo',loop_externo,'tempo do loop',time() - ti,'tempo total',tempoTotal)
```
