```python
exemplos = [estimativa]
def analisar_estimativa(estimativa):
    # Converter todas as palavras para minsculas
    estimativa = [palavra.lower() for palavra in estimativa]

    # Contar quantas palavras aparecem exatamente 1 vez
    contagem_uma_vez = {palavra: estimativa.count(palavra) for palavra in set(estimativa) if estimativa.count(palavra) == 1}

    # Contar quantas palavras aparecem exatamente 2 vezes
    contagem_duas_vezes = {palavra: estimativa.count(palavra) for palavra in set(estimativa) if estimativa.count(palavra) == 2}

    # Contar quantas palavras distintas existem
    palavras_distintas = list(set(estimativa))
    
    return {
        "contagem_uma_vez": len(contagem_uma_vez),
        "contagem_duas_vezes": len(contagem_duas_vezes),
        "palavras_distintas": len(palavras_distintas)
    }

# Exemplos de teste


for idx, exemplo in enumerate(exemplos):
    resultado = analisar_estimativa(exemplo)
    print(f"Exemplo {idx + 1}:")
    print(f"Palavras que aparecem uma vez: {resultado['contagem_uma_vez']}")
    print(f"Palavras que aparecem duas vezes: {resultado['contagem_duas_vezes']}")
    print(f"Palavras distintas: {resultado['palavras_distintas']}")
    print()
```
